```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("dataset.csv", sep=";").values
# print(dataset)
```

```python
survived = np.array([value[0] for value in dataset if value[1] == 'survived'])

plt.figure(figsize=(15,5))
plt.plot(survived, marker = 'o')
plt.grid(True)
plt.title('Survived sparrows')
plt.ylabel('Weight')
plt.xlabel('Index')
plt.show()

print("Dataset of survived sparrows:\n")
print(survived)
print("\n\nStatistical estimations of survived sparrows:\n")
print("\t- Expectation: ", np.average(survived))
print("\t- Variance:    ", np.var(survived))
print("\t- Median:      ", np.median(survived))
```

```python
perished = np.array([value[0] for value in dataset if value[1] == 'perished'])

plt.figure(figsize=(15,5))
plt.plot(perished, marker = 'o')
plt.grid(True)
plt.title('Perished sparrows')
plt.ylabel('Weight')
plt.xlabel('Index')
plt.show()

print("Dataset of perished sparrows:\n")
print(perished)
print("\n\nStatistical estimations of perished sparrows:\n")
print("\t- Expectation: ", np.average(perished))
print("\t- Variance:    ", np.var(perished))
print("\t- Median:      ", np.median(perished))
```
